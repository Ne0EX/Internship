```python
n = int(input("Input amount of data: "))                                                          # Receive  amount of value
x = ''

while n>0:
    
    alpha = input()                                                                                         # Input Strings
    alpha =  alpha.split()                                                                                 # Split with space
    
    for i in alpha:
        if i[0].isupper() == True:                                                                      #Check UpperCase letter in first alphabet
            x += i[0]
    
    print(x)                                                                                                     #Show result
    x = ''
    
    n -=1
```

```python
import math                                                                                                    # Import for use square root

mode = 0                                                                                                         # Declare variables
m = 1
j = 2

while True:                                                                                                     # Loop for recieve value which not equal 0.0
    
    n = float(input("Input amount of data: "))                                                        # Receive value
    
    if n==0.0: break                                                                                          # Value must not equal 0 
        

    if n<1 or n>10:                                                                                            # Check range of Value
        print(False)
    
    if (n*(10**11)) > 10**12 :                                                                           # Check digit of float
        print(False)
    
    if mode == 0:
        
        for i in range(1,4):                                                                                     #Loop for check 3 digit of float         
            if mode == 1: break
            a = int(n*10**i)
            

            while j<math.sqrt(a):                                                                             #Loop for find floating prime 
                if a%j == 0: break
                if a%j != 0 and j==(int(math.sqrt(a))-1):
                    
                    mode = 1
                    j = 2
                    break
                j+=1

            j = 2                                                                                                       #Set default
            
            
    if mode == 1:                                                                                                 #Show result
        print(True)
        mode = 0
        
    else: print(False)
        

            
    
```

```python
Hangman = input()

default = ''

result = ['_','_','_','_','_','_','_','_','_','_','_','_']

if ' ' not in Hangman:
    raise ValueError('You must space between each number')
    
Hangman = Hangman.split(' ')
Hangman = list(map(int,Hangman))

if len(Hangman) != 12:
    raise ValueError('You must enter 12 value for playing Hangman games')

if any(y<0 or y > 10 for y in Hangman):
    raise ValueError('Value must between 0 to 9 only!')

    
for i in range(0,5):
    predict = int(input())
    
    for i in range(0,12):
        if predict == Hangman[i]:
            result[i] = predict
            
            
    for i in range(0,len(Hangman)):
        default += str(result[i]) + ' '
            
    print(default)
    default = ''
    
score = 12 - result.count('_')
print(score)
```
